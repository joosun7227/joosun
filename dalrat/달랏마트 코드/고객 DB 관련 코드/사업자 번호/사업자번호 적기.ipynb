```python
import pandas as pd
```

```python
customer_df=pd.read_csv("export_customer_20240418_070904.csv")
business_df=pd.read_excel("사업자DB(new) (1).xlsx")
```

```python
customer_df
```

```python
business_df
```

```python
# Step 1: Add a leading zero to 'mobile_number' in export_customer data
customer_df['mobile_number'] = '0' + customer_df['mobile_number'].astype(str)

# Step 2: Remove dashes and concatenate the numbers in '휴대폰 번호' in business_db_data
business_df['휴대폰 번호'] = business_df['휴대폰 번호'].str.replace('-', '').astype(str)

# Step 3: Merge the two datasets on the modified phone number columns and get business registration numbers
merged_data = pd.merge(
    customer_df, 
    business_df, 
    left_on='mobile_number', 
    right_on='휴대폰 번호', 
    how='left'
)

# Select relevant columns for display
result_data = merged_data[['mobile_number', '사업자 등록번호']]
result_data.head()
```
