```python
import pandas as pd
import re
```

```python
# 파일 불러오기
customer_df = pd.read_csv('export_customer_20240324_231630.csv')
business_df = pd.read_excel('(구)달랏마트 대리점리스트.xls.xlsx')
```

```python
customer_df.info()
```

```python
business_df.info()
```

```python
# Step 1 & 2: Add '0' at the beginning if missing and convert 'mobile_number' to string
customer_df['mobile_number'] = customer_df['mobile_number'].apply(lambda x: '0' + str(x) if not str(x).startswith('0') else str(x))

# Step 3: Remove '-' from '연락처' in business_df and keep the numbers only
business_df['연락처'] = business_df['연락처'].str.replace('-', '', regex=False)

# Step 4: Merge customer_df and business_df based on 'mobile_number' and '연락처'
merged_df = pd.merge(customer_df, business_df, left_on='mobile_number', right_on='연락처', how='left')

merged_df.head()
```

```python
merged_df.info()
```

```python
# Selecting only the specified columns from the adjusted merged dataframe
merged_df = merged_df[['email', 'firstname', 'lastname', 'mobile_number', '아이디', '이름', '연락처']].copy()

merged_df.head()
```

```python
# 결과 확인 및 파일 다시 저장
output_path = 'jiyhean_code.xlsx' #저장 이름 및 경로 지정
merged_df.to_excel(output_path, index=False) #인덱스 포함하지 않고 저장
```
