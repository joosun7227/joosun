```python
grade=[]

for avg in df2.평균 :
    if avg >= 90:
        grade.append('A')
    elif avg >= 80:
        grade.append('B')
    else:
        grade.append('c')
```

```python
df2['성적']=grade
```

```python
df2
```

```python
def scholarship(row): 
    if row == "A": 
        return '장학생'
    else:
        return '비장학생'
df2['장학생'] = df2.성적.apply(scholarship)
df2
```
